### Durée minimale par trajet

Ici, j'ai d'abord regroupé d'abord les données avec o_city et d_city. Ensuite, j'ai parcouru les données avec une boucle for afin de calculé la durée en faisant la différence entre l'horaire d'arrivée et l'horaire de départ.

```python
#Durée min
res = ticket_data.groupby(['o_city','d_city'])
for a in res:
    print(a[0])
    print('----------------------')
    print(min(pd.to_datetime(a[1]['arrival_ts']) - pd.to_datetime(a[1]['departure_ts'])))
```

### Prix maximal par trajet

```python
#Durée max
res = ticket_data.groupby(['o_city','d_city'])
for a in res:
    print(a[0])
    print('----------------------')
    print(max(pd.to_datetime(a[1]['arrival_ts']) - pd.to_datetime(a[1]['departure_ts'])))
```

### Durée moyenne par trajet

```python
#Durée moy
from statistics import mean
res = ticket_data.groupby(['o_city','d_city'])
for a in res:
    print(a[0])
    #print(a)
    print('----------------------')
    print(mean((pd.to_datetime(a[1]['arrival_ts']) - pd.to_datetime(a[1]['departure_ts'])).astype('timedelta64[m]')))
```

## Différence de prix moyen et durée selon le train, le bus et le covoit selon la distance du trajet (Par exemple (0-200km, 201-800km, 800-2000km, 2000+km))

Dans cette partie, je vais d'abord calculer la distance entre deux villes grâce aux latitudes et longitudes données pour ensuite finir par grouper le prix moyen et la durée du trajet en fonction du type de transport et de cette distance

```python
q5 = "SELECT o_city, cities.latitude, cities.longitude FROM cities Join ticket_data on ticket_data.o_city=cities.id"
sqldf(q5, globals())
```

```python
#création de listes contenant les latitudes et longitudes des villes de départ et d'arrivée
list_lat1 = []
list_long1 = []
list_lat2 = []
list_long2 = []

for i in ticket_data['o_city']:
    #print(i)
    #print(cities[(cities['id']==76)].latitude)
    #print(cities[(cities['id']==76)].longitude)
    list_lat1.append(cities[(cities['id']==i)].latitude)
    list_long1.append(cities[(cities['id']==i)].longitude)
    
#print(list_lat1[0])
for j in ticket_data['d_city']:
    list_lat2.append(cities[(cities['id']==j)].latitude)
    list_long2.append(cities[(cities['id']==j)].longitude)
```

```python
#Calcul de la distance

from math import sin, cos, sqrt, atan2, radians
R = 6373.0

list_kilometre = []
for i in range(0, len(list_lat1)):
    lon1 = list_long1[i]
    lon2 = list_long2[i]
    lat1 = list_lat1[i]
    lat2 = list_lat2[i]
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2*atan2(sqrt(a), sqrt(1-a))

    list_kilometre.append(R*c)
```
